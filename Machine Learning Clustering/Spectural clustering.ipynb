```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x= dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import SpectralClustering


sc = SpectralClustering(n_clusters=4,assign_labels='discretize',random_state=0).fit(x)
labels = sc.labels_

# plt.plot(range(0,11),cluster_centers_indices)
# plt.xlabel('number of clusters')
# plt.ylabel('WCSS')
# plt.show()
```

```python
SpectralClustering(assign_labels='discretize', n_clusters=2,
    random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1], c=labels)
plt.show() 
```

```python
 
f = plt.figure()
f.add_subplot(2, 2, 1)
for i in range(2, 6):
 sc = SpectralClustering(n_clusters=i).fit(x)
 f.add_subplot(2, 2, i-1)
 plt.scatter(x[:,0], x[:,1], s=5, c=sc.labels_, label="n_cluster-"+str(i))
 plt.legend()

plt.show() 
```
